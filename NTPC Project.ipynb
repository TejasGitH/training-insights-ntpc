```python
import pandas as pd
import numpy as np
# Show all rows
pd.set_option('display.max_rows', None)

# Show all columns
pd.set_option('display.max_columns', None)
```

```python
# Replace 'your_file.xlsx' with the actual path to your file
df = pd.read_excel('NTPC.xlsx', engine='openpyxl')
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
list(df.columns)
```

```python
df
```

```python
df.drop('New Programme', axis=1, inplace=True)
```

```python
df.drop('Non Exec. Programme', axis=1, inplace=True)
```

```python
df.drop('Training Coordinator Email-ID', axis=1, inplace=True)
```

```python
df.drop('Competency Based ', axis=1, inplace=True)
```

```python
df = df.rename(columns={'Competeny\nBased': 'Competency Based'})
```

```python
df['Scheduled/Unscheduled/Drum/Sponsored'] = df['Scheduled/Unscheduled/Drum/Sponsored'].fillna('Unscheduled')
```

```python
df['Residential/Non-Residential'] = df['Residential/Non-Residential'].fillna('Non-Residential')
```

```python
df = df.drop(df.index[-1])
```

```python
df
```

```python
df.to_csv('Cleaned_NTPC.csv' , index = False)
```

### 1)Count of Each 'Programme Name - Abbreviation'

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count occurrences
top_programmes = df['Programme Name - Abbreviation'].value_counts().head(15)

# Plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_programmes.values, y=top_programmes.index, hue=top_programmes.index)

# Add count labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=12)
    
plt.title('Top 15 Most Frequent Programme Abbreviations')
plt.xlabel('Count')
plt.ylabel('Programme Abbreviation')
plt.tight_layout()
plt.show()
```

### 2)Pie Chart of Participant Types (Internal vs External vs Both)

```python
participant_counts = df['Participant (Internal (I)/External (E))BOTH /I+E'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(participant_counts, labels=participant_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Participant Profile: Internal vs External vs Both')
plt.tight_layout()
plt.show()
```

### 3)Program Category by Residential Status

```python
pivot_data = df.pivot_table(index='Category   (Tech (T) / Mgmt (M)/  Leadership (L))', 
                             columns='Residential/Non-Residential', 
                             values='SN', aggfunc='count').fillna(0)

ax = pivot_data.plot(kind='bar', stacked=False, figsize=(10, 6), colormap='tab20c')

# Add count labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=12)
    
plt.title('Program Category by Residential Status')
plt.xlabel('Category')
plt.ylabel('Number of Programs')
plt.tight_layout()
plt.show()
```

### 4)Distribution of program durations

```python
duration_counts = df['Duration'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=duration_counts.index, y=duration_counts.values, color='teal')
# Add count labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=12)
plt.title('Number of Programs by Duration')
plt.xlabel('Duration (Days)')
plt.ylabel('Number of Programs')
plt.tight_layout()
plt.show()
```

### 5)Top training categories

### 15)Programs per Month (Seasonality) in 2024 and 2025

```python
# Re-import necessary libraries after code state reset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Start Date column to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

# Filter for year 2024 only
df_2024 = df[df['Start Date'].dt.year == 2024].copy()

# Extract month names
df_2024['Month'] = df_2024['Start Date'].dt.month_name()

# Define month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

# Plot
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df_2024, x='Month', order=month_order, hue = 'Month')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=12)
plt.title('Training Programs Conducted by Month (2024)')
plt.xlabel('Month')
plt.ylabel('Number of Programs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
